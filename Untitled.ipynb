```python
import pandas as pd
df = pd.ExcelFile('gesamtdatei_stundenwerte.xlsx')
df = pd.read_excel(df, 'Jahresdatei 2021')
df = pd.DataFrame(df)
df
```

```python
df['Datetime']= pd.to_datetime(df['Date']+' '+df['Time'])
df.drop(['Date', 'Time'], axis=1, inplace= True)
df.set_index('Datetime', inplace= True)
print(df.resample('H').sum(min_count=1))

# output
                       Value
Datetime                    
2021-11-06 04:00:00  12637.1
2021-11-06 05:00:00  10859.4
2021-11-06 06:00:00      NaN
2021-11-06 07:00:00   1111.1
```
